```python
#import xarray as xr
from earthkit.regrid import interpolate
from earthkit.data import from_source
```

```python
#ds = xr.open_dataset('', engine='cfgrib')
ds = from_source('file', 'pr_ECMWF_O640_2025-02_globe.grib')
```

```python
ds
```

```python
out_grid = {'grid': [0.125, 0.125]}
```

```python
r = interpolate(ds, out_grid=out_grid, method="linear")
```

```python
d = r.data()
lat = d[0]
lon = d[1]
vals = d[2:]
```
